```python
df = pd.read_csv("lista.csv", sep=",")
```

```python
lista = list(df.groupby(by= ["country"]).sum().sort_values("cases", ascending=False).reset_index().country)
```

```python
df.groupby(by= ["country"]).sum().sort_values("cases", ascending=False).reset_index().country
```

```python
lista
```

```python
def sostituzione(elemento):
    dizionario = {"Gen":1, "Feb":2, "Mar":3, "Apr":4, "Mag":5, "Giu":6, "Lug":7, "Ago":8, "Set":9,
                   "Ott":10, "Nov":11, "Dic":12}
    return dizionario[elemento]
df["month_num"] = df.month.apply(sostituzione)
```

```python
df["country_num"] = df.country.apply(lambda x: lista.index(x))
```

```python
df
```

```python
lista
```

```python
df.columns
```

```python
dfs = []
for mese in range(1,13):
    if(len(df.loc[df.month_num==mese]!= 10)):
        paesi = list(df.loc[df.month_num==mese].country)
        mese_stringa = pd.unique(df.loc[df.month_num == mese].month)
        for paese in lista:
            if(paese not in paesi):
                dfs.append(pd.DataFrame({"month": mese_stringa, "country": paese, "cases":0, "deaths":0, "month_num": mese,
                                         "country_num": lista.index(paese)}))
```

```python
for riga in dfs:
    df = pd.concat([riga, df])
```

```python
df
```

```python
df = df.sort_values(by=["month_num",  "deaths"], ascending=False).reset_index(drop=True)
```
